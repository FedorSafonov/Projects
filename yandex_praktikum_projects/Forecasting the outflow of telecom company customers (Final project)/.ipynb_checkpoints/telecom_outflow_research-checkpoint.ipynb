# Описание

**Задача бизнеса**: Отдел маркетинга «Ниединогоразрыва.ком» приступает к проработке способов иудержания клиентов и уменьшения оттока.

**Цель проекта:** исследовать распределение и статистики двух категорий клиентов - тех кто остался, и тех кто ушёл.

**Данные:** Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

**Описание данных:**

Данные состоят из файлов, полученных из разных источников:

- `contract.csv` — информация о договоре;
- `personal.csv` — персональные данные клиента;
- `internet.csv` — информация об интернет-услугах;
- `phone.csv` — информация об услугах телефонии.

Во всех файлах столбец `customerID` содержит код клиента.

Информация о договорах актуальна на 1 февраля 2020.

**Описание услуг:**

Оператор предоставляет два основных типа услуг: 

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL*,* от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).  

Также доступны такие услуги:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

# Первичная предобработка

Так как данные уже были проверены на дубликаты, пропуски и аномалии в основном проекте и никаких проблем обнаружено не было, то здесь я пропущу этот этап.

```python
def to_float(df,_):
    '''Функция преобразует формат столбца в минимально подходящий формат float
    на вход подаётся название датафрейма и столбца.'''
    df[_] = pd.to_numeric(df[_], downcast='float',errors='coerce')
    
def to_uint(df,_):
    '''Функция преобразует формат столбца в минимально подходящий формат uint
    на вход подаётся название датафрейма и столбца.'''
    df[_] = pd.to_numeric(df[_], downcast='unsigned',errors='coerce')

def to_int(df,_):
    '''Функция преобразует формат столбца в минимально подходящий формат int
    на вход подаётся название датафрейма и столбца.'''
    df[_] = pd.to_numeric(df[_], downcast='integer',errors='coerce')
```

## Типы данных

Приведу типы в правильный формат, добавлю целевой признак и lifetime клиента, чтобы их тоже можно было исследовать. Булевый формат 'Yes'-'No' приведу в 1-0 для удобста.

```python
contract["has_client_left"] = contract.apply(lambda row: 0 if row["EndDate"] == 'No' else 1, axis=1)
for i in ['EndDate','BeginDate']: 
    contract[i] = pd.to_datetime(contract[i]
                                .replace('No','2020-02-01'),
                                 format = '%Y-%m-%d')
contract['lifetime'] = (contract['EndDate'] - contract['BeginDate']).dt.days
contract['TotalCharges'] = contract['TotalCharges'].fillna(0)
contract['PaperlessBilling'] = contract['PaperlessBilling'].isin(['Yes'])*1

for i in ['TotalCharges','MonthlyCharges']: to_float(contract, i)
for i in ['lifetime','PaperlessBilling','has_client_left']: to_uint(contract, i)

personal[['Partner', 'Dependents']] = personal[['Partner', 'Dependents']].isin(['Yes'])*1
for i in personal.columns[2:]: to_uint(personal,i)
    
phone['MultipleLines'] = phone['MultipleLines'].isin(['Yes'])*1
to_uint(phone,'MultipleLines')

columns = ['OnlineSecurity', 'OnlineBackup', 
          'DeviceProtection', 'TechSupport',
          'StreamingTV', 'StreamingMovies']
internet[columns] = (internet[columns]).isin(['Yes'])*1
for i in internet.columns[2:]: to_uint(internet,i)
```

## Соединение таблиц

```python
df = contract.merge(personal, how='left').merge(phone, how='left').merge(internet, how='left')
```

```python
if (df['MultipleLines'].count() == phone['MultipleLines'].count()
and df['lifetime'].count() == contract['lifetime'].count()
and df['StreamingTV'].count() == internet['StreamingTV'].count()):
    print('Данные обьеденены верно')
else:
    print('Данные перенесены некорректно')
```

Заполню пропуски значением -1, чтобы показать, что клиент не пользуется этим видом связи.

```python
df = df.fillna(-1)
columns.append('MultipleLines')
for i in columns: to_int(df,i)
```

Выделю две выборки - оставшихся и ушежших клиентов.

```python
remained = df[df['has_client_left']==0]
outflow = df[df['has_client_left']==1]
```

# Графики распределения

```python
data = {'всех': df, 'оставшихся': remained, 'ушедших': outflow}
colors = ['indigo', 'chartreuse', 'chocolate']

for i, (key, df) in enumerate(data.items()):
    df['MonthlyCharges'].plot.hist(bins=50, color=colors[i],figsize=(8,3))
    
    mean_line = df['MonthlyCharges'].mean()
    plt.plot([mean_line, mean_line], [0, 900], '--', color='darkblue', label='Среднее')
    
    median_line = df['MonthlyCharges'].median()
    plt.plot([median_line, median_line], [0, 900], '--', color='darkgreen', label='Медиана')
    
    q25, q75 = df['MonthlyCharges'].quantile([0.25, 0.75])
    plt.plot([q25, q25], [0, 900], '--', color='red', label='25%-квартиль')
    plt.plot([q75, q75], [0, 900], '--', color='red', label='75%-квартиль')
    
    plt.ylim(0,900)
    plt.title(f'График распределение ежемесячного платежа {key} клиентов')
    plt.ylabel('Количество клиентов',size=15)
    plt.xlabel('Ежемесячный платёж',size=15)
    plt.legend()
    plt.show()
```

- На графике видно, что уходящие клиенты больше распределены в диапазоне от 65 до 110.
- А оставшиеся наоборот многие имеют низкий платеж от 20 до 30, а затем распределяться в диапазоне от 40 до 115

# Доля пользователей телефонной связи и интернет-услуг в разных категориях

```python
fig, ax = plt.subplots(nrows=3, ncols=2)

for i, (key, df) in enumerate(data.items()):
    for j, column in enumerate(['MultipleLines', 'StreamingTV']):
        subplot = ax[i][j]
        
        mask = df[column] == -1
        num_negative_values = df[mask].shape[0]
        num_other_values = df.shape[0] - num_negative_values
        labels = ['Не пользутся услугами', 'Пользуются услугами']
        values = [num_negative_values, num_other_values]
        subplot.bar(labels, values, color=colors[i])
        for k, v in enumerate(values):
            subplot.text(k, v + 100, f'{v / df.shape[0]:.2%}', ha='center')
        if column == 'MultipleLines':
            subplot.set_title(f'Доля пользователей телефонной связи среди {key} клиентов',size=8)
        else:
            subplot.set_title(f'Доля интернет-пользователей среди {key} клиентов',size=8)
        subplot.set_ylim(0,df.shape[0]+200)

plt.tight_layout()
fig.set_size_inches(10, 9)
fig.text(0.01, 0.5, 'Количество пользователей', va='center', rotation='vertical', size=30)  
fig.subplots_adjust(top=0.965,hspace=0.3,wspace=0.3)
plt.show()
```

- Доля пользователей телефонной связи не сильно отличается - 9.1% среди оставшихся и 9.9% среди ушедших не пользуются услугами.
- А вот в доля интернет-пользователей отличается сильнее.  Среди ушедших клиентов 94% клиентов пользовались интрнетом, а вот среди оставшихся уже всего 72.7%
