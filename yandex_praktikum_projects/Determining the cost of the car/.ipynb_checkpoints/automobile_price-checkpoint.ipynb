```python
# Код для удобного оформления предобработки.

def create_cell(shell, text):
    payload = dict(
        source='set_next_input',
        text=text,
        replace=False,
    )
    shell.payload_manager.write_payload(payload, single=False)

def column_preprocess_description(df):
    shell = get_ipython()
    for i, column in reversed(list(enumerate(df.columns))):
        create_cell(shell, dedent(
            '''
            * Изменение
            '''
        ))
        create_cell(shell, dedent(
            '''
            * Вывод
            \t- Обнаруженные аномалии: ...
            \t- Причины, по которым они могли возникнуть: ...
            \t- Как исправлять: ...
            \t- Финальный тип и почему: ...
            '''
        ))
        create_cell(shell, dedent(
            f'''
            investigate_column(df["{column}"])
            '''
        ))
        create_cell(shell, dedent(
            '''
            * Изучим значения
            '''
        ))
        create_cell(shell, dedent(
            f'''
            #### {i+1}. Обработка "{column}"
            '''
        ))
```

```python
def investigate_column(column):
    print('Fraction of Null or Na is', np.round(column.isna().mean(),5))
    print(column.describe())
```

* Изучение

Создадим функцию для группировки данных, она понадобиться для изучения данных, а так же оценки столбца с точки зрения информативности и полезности для обучения модели.

```python
def group(df,column,ags,n):
    '''Функция берёт на вход датафрейм, название столбца, действий 
    и номер действия для сортировки.
    Возвращает сгруппированный датафрей по целевому признаку "Price"'''
    return df.groupby(column)['Price'].agg(ags).sort_values(ags[n],ascending=False)
```

## Обучение моделей

* Делим признаки на целевой и остальные.

```python
X = dfg.drop('Price',axis=1)
y = dfg['Price']
```

* Разбиваем выборку на тестову и обучающая (валидационная не понадобиться потому что я буду использовать кросс-валидацию)

```python
X_train, X_test, y_train, y_test = tts(X, y, test_size = 0.25, random_state = s)
```

* Выделяем категоричальные признаки и создаем основу для OHE-кодирования и OE-кодирования. OHE понадобиться нам для обучения линейных моделей (Линейная Регрессия), а OE для бустингов и деревьев.

```python
cat_features = ['VehicleType', 'Gearbox', 'Model', 'FuelType', 'Brand', 'NotRepaired']

ohe_transformer = make_column_transformer(
    (OneHotEncoder(), cat_features),
    remainder='passthrough')

oe_transformer = make_column_transformer(
    (OrdinalEncoder(), cat_features),
    remainder='passthrough')
```

<div class="alert alert-block alert-success">

<b>Успех:</b> Отлично, что используются рзаные кодировщики для разных моделей. С OE древовидные модели и бустинги будут работать эффективнее с точки зрения времени, а для заказчика это важно.

* Создаем датафрейм в который будем заносить результаты работы моделей для сравнения.

```python
results = pd.DataFrame(columns=['RMSE','Время обучения','Время предсказания'])
results
```

```python
# Добавляем опцию, чтобы все дробные числа в таблицах
# округлялись до второго знака после запятой.

pd.options.display.float_format = '{: .2f}'.format
```

* Создаем функцию для занесенися результатов в таблицу.

```python
def metrics(df,model,RMSE,fit,pred):
    '''Функция берёт на вход датафрейма, название модели, показаетль RMSE (по CV),
    время обучения и время предсказания. Добавляет строку с показателями модели.'''
    
    df.loc[model,'Время обучения'] = fit
    df.loc[model,'Время предсказания'] = pred
    df.loc[model,'RMSE'] = RMSE
    
    return df
```

* Создаем функцию которая будет производить обучение модели, делать предсказания и фиксировать время этих процессов.

<div class="alert alert-block alert-warning">
<b>Изменения:</b> Были внесены следующие изменения:
    Исправил функцию, теперь время предсказания достаётся из mean_test_score и тестовая выборка не трогается.
    Изменил вариант else - теперь он нужен для проверки лучше модели на тестовой выборке.
</div>

<div class="alert alert-block alert-success">
<b>Успех[2]:</b> Отлично!

```python
def timer(model,X_train,y_train=None,gs=False,params=None):
    """Фукция берет на вход модель, обучающие выборки с признаком и таргетом,
    если подаем Grid Search, то указываем gs=True и добавляем параметры для него."""
    
    if gs == True:
        grid = GridSearchCV(model, params, scoring='neg_root_mean_squared_error')
        grid.fit(X_train,y_train)
        i = (list(grid.cv_results_['rank_test_score'])).index(1)
        time_fit = grid.cv_results_['mean_fit_time'][i]
        print('Лучшие параметры модели:', grid.best_params_)
        rmse = (grid.cv_results_['mean_test_score'][i]*-1)
        print('RMSE',rmse)
        print('Время обучения модели:', time_fit)
        time_pred = grid.cv_results_['mean_score_time'][i]
        print('Время предсказания модели:', time_pred)
        
    else:    
        start = time.time()
        model.fit(X_train, y_train)
        end = time.time()
        time_fit = end - start
        grid = None

        print('Время обучения модели:', time_fit)
        start = time.time()
        preds = model.predict(X_test)
        end = time.time()
        rmse = mse(y_test, preds) ** 0.5
        time_pred = end - start
        print('Время предсказания модели:', time_pred)
        print('RMSE',rmse)
    
    
    return grid, time_fit, time_pred, rmse
```
