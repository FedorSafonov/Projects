### Эффективность обогощения (проверка рассчётов)

Вначале познакомимся с самой формулой

**recovery** = ((C(F-T))/(FC-T))*100%

* **С** - доля золота концентрате после фильтрации/очитски
* **F** - доля золота в сырье/концентрате до флотации/очистки
* **T** - доля золота в отвальных хвостах после фильтрации/очистки

```python
def recovery(df,C,F,T):
    '''Функция рассчитывает эффективность обогощения и сравнивает с истинными значениями
    на экран выводиться показатель MAE и среднее recovery по выборке'''
    
    recovery = ((df[C]*(df[F]-df[T]))/(df[F]*(df[C]-df[T]))) * 100
    MAE = mae(df['rougher.output.recovery'],recovery)
    print('MAE:{0:.16f}'.format(MAE))
    print('recovery_mean:{0:.2f}'.format(recovery.mean()))
    return MAE
```

```python
C = 'rougher.output.concentrate_au'
F = 'rougher.input.feed_au'
T = 'rougher.output.tail_au'

recovery(train,C,F,T)
```

Показатель МАЕ, который показывает среднее рассхождение предсказаний от истинных значений стремиться к нулю. Значит с рассчётами всё в порядке

## Модель

### Функция sMAPE

```python
def sMAPE(target, predict):
    sMAPE= 1/len(target) * np.sum(2 * np.abs(target-predict) / (np.abs(target) + np.abs(predict))*100)
    return sMAPE
```

```python
def final_sMAPE(rougher, final):
    final = 0.25*rougher+0.75*final
    return final
```

```python
smape_scorer = make_scorer(sMAPE, greater_is_better=False)
```

```python
train = train.sample(frac=1).reset_index(drop=True)
test = test.sample(frac=1).reset_index(drop=True)
```

```python
#features
f_train = train.drop(['rougher.output.recovery','final.output.recovery'],axis=1) 
tr_train = train['rougher.output.recovery']         # target rougher
tf_train = train['final.output.recovery']   # target final                     
```

```python
def cross_and_smape_r(model):
    '''Функция производит кросс-валидацию выбранной модели для целевого признака 'rougher.output.recovery' 
    и выводит средний показатель sMAPE'''
    cross = cv_s(model,
                        f_train,
                        tr_train,
                        scoring=smape_scorer,
                        cv=5,n_jobs=-1)
    return abs(np.mean(cross))

def cross_and_smape_f(model):
    '''Функция производит кросс-валидацию выбранной модели для целевого признака 'final.output.recovery'
    и выводит средний показатель sMAPE'''
    cross = cv_s(model,
                        f_train,
                        tf_train,
                        scoring=smape_scorer,
                        cv=5,n_jobs=-1)
    return abs(np.mean(cross))
```

```python
def GridSearch_R(model,params):
    '''Функция производит подбор лучших параметров выбранной модели для целевого признака 'rougher.output.recovery' 
    и выводит средний показатель sMAPE'''
    GSCV_R = GridSearchCV(model,params,scoring=smape_scorer,
                        cv=5,n_jobs=-1)
    GSCV_R.fit(f_train,tr_train)
    smape_r = abs(np.mean(GSCV_R.cv_results_['mean_test_score']))
    return  smape_r, GSCV_R.best_estimator_ 
    

def GridSearch_F(model,params):
    '''Функция производит подбор лучших параметров выбранной модели для целевого признака 'final.output.recovery' 
    и выводит средний показатель sMAPE'''
    GSCV_F = GridSearchCV(model,params,scoring=smape_scorer,
                        cv=5,n_jobs=-1)
    GSCV_F.fit(f_train,tf_train)
    smape_r = abs(np.mean(GSCV_F.cv_results_['mean_test_score']))
    return  smape_r, GSCV_F.best_estimator_ 
     
```

```python
def GridSearch_sMAPE(model,params,model_name):
    smape_rougher,gs_model_r = GridSearch_R(model,params)
    smape_final,gs_model_f = GridSearch_F(model,params)
    f_smape = smape_rougher*0.25 + smape_final*0.75
    
    print('sMAPE rougher для модели {} равно {:.2f}'.format(model_name,smape_rougher))
    print('Параметры:',gs_model_r)
    print()
    print('sMAPE final для модели {} равно {:.2f}'.format(model_name,smape_final))
    print('Параметры', gs_model_f)
    print()
    print('Итоговое sMAPE для модели {} равно {:.2f}'.format(model_name,f_smape))
    print()
```

```python
def valid_sMAPE(model,model_name):
    smape_rougher = cross_and_smape_r(model)
    smape_final = cross_and_smape_f(model)
    f_smape = smape_rougher*0.25 + smape_final*0.75
    
    print('sMAPE rougher для модели {} равно {:.2f}'.format(model_name,smape_rougher))
    print('sMAPE final для модели {} равно {:.2f}'.format(model_name,smape_final))
    print('Итоговое sMAPE для модели {} равно {:.2f}'.format(model_name,f_smape))
    print()
```

### Выбор модели

```python
%%time
model_lr = LR()
parametrs = { 'normalize':[True,False] }
GridSearch_sMAPE(model_lr,parametrs,'LinearRegression')
```

```python
%%time
model_RFR = RFR(min_samples_leaf=7,min_samples_split=5)
parametrs = { 'n_estimators': [50, 100],
              'max_depth': range (4,10,2),}
GridSearch_sMAPE(model_RFR,parametrs,'Случайный лес')
```

```python
%%time

model_CBR = CBR(iterations=1000,
                depth=7,
                logging_level='Silent',
                )

valid_sMAPE(model_CBR,'CatBoost')
```

Самой лучшей моделью вышел CatBoostRegressor, который даже без перебора параметров оказался лучше леса и регрессии. Поэтому именно его будем использовать на тестовой выборке.

### Проверка модели на тестовой выборке

```python
#features
f_test = test.drop(['rougher.output.recovery','final.output.recovery'],axis=1) 
tr_test = test['rougher.output.recovery']         # test rougher
tf_test = test['final.output.recovery']   # test final                     
```

```python
%%time

CBR_r = CBR(iterations=1000,
                depth=7,
                logging_level='Silent',
                )
CBR_f = CBR(iterations=1000,
                depth=7,
                logging_level='Silent',
                )

CBR_r.fit(f_train, tr_train)
CBR_f.fit(f_train, tf_train)

CBR_r_pred = CBR_r.predict(f_test)
CBR_f_pred = CBR_f.predict(f_test)

rougher = sMAPE(tr_test, CBR_r_pred)
final = sMAPE(tf_test, CBR_f_pred)

final_smape = final_sMAPE(rougher, final)
```

Для проверки модели на адекватность сравним её со случайной моделью, которая всегда угадывает среднее число

```python
dummy_r = DummyRegressor(strategy="mean")
dummy_f = DummyRegressor(strategy="mean")

dummy_r.fit(f_train, tr_train)
dummy_f.fit(f_train, tf_train)

dummy_pred_r = dummy_r.predict(f_test)
dummy_pred_f = dummy_f.predict(f_test)

rougher_d = sMAPE(tr_test, dummy_pred_r)
final_d = sMAPE(tf_test, dummy_pred_f)

final_smape_dummy = final_sMAPE(rougher_d, final_d)
```
