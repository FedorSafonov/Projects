# Отток клиентов

**Описание** - Из «Бета-Банка» стали уходить клиенты. Каждый месяц. Немного, но заметно. Банковские маркетологи посчитали: сохранять текущих клиентов дешевле, чем привлекать новых.

**Цель** - Нужно спрогнозировать, уйдёт клиент из банка в ближайшее время или нет.

**Данные** - Исторические данные о поведении клиентов и расторжении договоров с банком. 

**Задача** - Постройте модель с предельно большим значением *F1*-меры. Чтобы сдать проект успешно, нужно довести метрику до 0.59. Проверьте *F1*-меру на тестовой выборке самостоятельно.

Дополнительно измеряйте *AUC-ROC*, сравнивайте её значение с *F1*-мерой.

Источник данных: [https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling](https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling)

## Исследование задачи

В начале сделаю метод OHE ('One hot encoder') для избавления от категориальных признаков

```python
df_ohe = pd.get_dummies(df, drop_first=True)
```

```python
#target - целевой признак
t = df_ohe['exited']

#features - признаки
f = df_ohe.drop(['exited'],axis=1) # Признаки
```

Так как у нас нет тестовой выборки я поделю датасет по пропорции 60-20-20 на обучающую, валидационную и тестувую выборки.

```python
f_train, f_valid, t_train, t_valid = t_t_s(f, t, test_size=0.4,random_state=s, stratify=t)

f_valid, f_test, t_valid, t_test = t_t_s(f_valid, t_valid, test_size=0.5,random_state=s, stratify=t_valid)

print(f_train.shape, f_valid.shape, f_test.shape)
print(t_train.shape, t_valid.shape, t_test.shape)
```

Выборки разделены, и пропорции сохранены верно. Следующим шагом проведу стандартизацию во всех выборках.

```python
scaler = StandardScaler()
numeric = ['creditscore', 'age', 'tenure', 'balance', 'numofproducts', 'estimatedsalary']
scaler.fit(f_train[numeric])
    
f_train[numeric] = scaler.transform(f_train[numeric])
f_valid[numeric] = scaler.transform(f_valid[numeric])
f_test[numeric] = scaler.transform(f_test[numeric])
```

```python
df['exited'].value_counts().plot.pie(legend=True,
                                     labeldistance=1.2,
                                     autopct='%1.2f%%',
                                     figsize=(12,6),
                                     title='Баланс классов',
                                     cmap='summer')
plt.show()
```

Баланс классов 20-80 (1к4), по условию задачи проведу обучение моделей без учёта дисбаланса классов.

### LogisticRegression

```python
%%time
solvers = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']

best_LR = None
best_solver_LR = None
best_f1_LR = 0
best_auc_roc_LR = 0


for sol in solvers:
    LR = LRC(solver=sol, max_iter=4000, random_state=s)
    LR.fit(f_train, t_train)
    pred_v_LR = LR.predict(f_valid)
    f1_LR = f1_s(pred_v_LR, t_valid)
    prob_valid_LR = LR.predict_proba(f_valid)
    prob_one_valid_LR = prob_valid_LR[:, 1]
    auc_roc_LR = roc_auc_score(t_valid, prob_one_valid_LR)
    print((f"{sol}: {f1_LR}"))
    if best_f1_LR < f1_LR:
            best_LR = LR 
            best_solver_LR = sol
            best_f1_LR = f1_LR
            best_auc_roc_LR = auc_roc_LR
print("______________________")
print('Лучший параметр solver:',best_solver_LR)
print('F1 Score лучшей модели Logistic Regression на валидационной выборке:',best_f1_LR)
print('AUC_ROC лучшей модели Logistic Regression на валидационной выборке:',best_auc_roc_LR)
print("______________________")
```

Разные solver'ы никак не отличаються в данном случае, f1 крайне низкий.

### DecisionTreeClassifier

### Downsampling

```python
def downsample(f, t, frac):
    '''Уменьшаем количество большего класса в выборке'''
    f_zeros = f[t == 0]
    f_ones = f[t == 1]
    t_zeros = t[t == 0]
    t_ones = t[t == 1]

    f_d = pd.concat([f_zeros.sample(frac=frac, random_state=s)] + [f_ones])
    t_d = pd.concat([t_zeros.sample(frac=frac, random_state=s)] + [t_ones])
    
    f_d, t_d = shuffle(f_d, t_d, random_state=s)
    
    return f_d, t_d

f_d, t_d = downsample(f_train, t_train, 0.25)

print(t_d.value_counts())
```

#### LogisticRegression

```python
%%time
solvers = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']

best_LR = None
best_solver_LR = None
best_f1_LR = 0
best_auc_roc_LR = 0


for sol in solvers:
    LR = LRC(solver=sol, max_iter=4000, random_state=s)
    LR.fit(f_d, t_d)
    pred_v_LR = LR.predict(f_valid)
    f1_LR = f1_s(pred_v_LR, t_valid)
    prob_valid_LR = LR.predict_proba(f_valid)
    prob_one_valid_LR = prob_valid_LR[:, 1]
    auc_roc_LR = roc_auc_score(t_valid, prob_one_valid_LR)
    print((f"{sol}: {f1_LR}"))
    if best_f1_LR < f1_LR:
            best_LR = LR 
            best_solver_LR = sol
            best_f1_LR = f1_LR
            best_auc_roc_LR = auc_roc_LR
print("______________________")
print('Лучший параметр solver:',best_solver_LR)
print('F1 Score лучшей модели Logistic Regression на валидационной выборке:',best_f1_LR)
print('AUC_ROC лучшей модели Logistic Regression на валидационной выборке:',best_auc_roc_LR)
print("______________________")
```

#### DecisionTreeClassifier

```python
best_DT = None
best_f1_DT = 0
best_depth_DT = 0
best_auc_roc_DT = 0

x=[]
y=[]
for depth in tqdm(range(1,15)):
    DT = DTC(random_state = s, max_depth = depth)
    DT.fit(f_d,t_d)
    pred_v_DT = DT.predict(f_valid)
    f1_DT = f1_s(pred_v_DT, t_valid)
    prob_valid_DT = DT.predict_proba(f_valid)
    prob_one_valid_DT = prob_valid_DT[:, 1]
    auc_roc_DT = roc_auc_score(t_valid, prob_one_valid_DT)
    x.append(depth)
    y.append(f1_DT)
    if best_f1_DT < f1_DT:
        best_DT = DT 
        best_f1_DT = f1_DT
        best_depth_DT = depth
        best_auc_roc_DT = auc_roc_DT
        
plt.figure(figsize=(12, 6))        
plt.plot(x, y)
plt.ylabel('F1 score',size=20)
plt.xlabel('Глубина',size=20)
plt.show()
        
print('Максимальная глубина наилучшей модели Decision Tree на валидационной выборке:', best_depth_DT)
print('F1 score лучшей модели DecisionTreeClassifier на валидационной выборке:',best_f1_DT)
print('AUC_ROC лучшей модели DecisionTreeClassifier на валидационной выборке:',best_auc_roc_DT)
print("______________________")
```

```python
%%time

best_RF = None
best_est_RF = 0
best_f1_RF = 0
best_depth_RF = 0
best_auc_roc_RF = 0

x=[]
y=[]

for depth in tqdm(range(1,25)):
    y1=[]
    for est in range(10,100,10):
        RF = RFC(random_state = s, max_depth = depth, n_estimators = est)
        RF.fit(f_d,t_d)
        pred_v_RF = RF.predict(f_valid)
        f1_RF = f1_s(pred_v_RF, t_valid)
        prob_valid_RF = RF.predict_proba(f_valid)
        prob_one_valid_RF = prob_valid_RF[:, 1]
        auc_roc_RF = roc_auc_score(t_valid, prob_one_valid_RF)
        y1.append(f1_RF)
        if best_f1_RF < f1_RF:
            best_RF = RF 
            best_est_RF = est
            best_f1_RF = f1_RF
            best_depth_RF = depth
            best_auc_roc_RF = auc_roc_RF
    x.append(depth)        
    y.append(max(y1))
    
plt.figure(figsize=(12, 6)) 
plt.plot(x, y)
plt.ylabel('Лучший F1-score каждого N деревьев',size = 15)
plt.xlabel('Глубина деревьев',size = 20)
plt.show()

print('Максимальная глубина наилучшей модели Random Forest на валидационной выборке:', best_depth_RF)
print('Количество деревьев наилучшей модели Random Forest на валидационной выборке:', best_est_RF)
print('F1 score лучшей модели RandomForestClassifier на валидационной выборке:',best_f1_RF)
print('AUC_ROC лучшей модели RandomForestClassifier на валидационной выборке:',best_auc_roc_RF)
print("______________________")
```
