–î–æ–±–∞–≤–∏–º –¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏ –∏ —á–∞—Å –≤ –∏—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ.

```python
df['dayofweek'] = df.index.dayofweek
df['hour'] = df.index.hour
```

–°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ–Ω–∞–¥–æ–±—è—Ç—Å—è –Ω–∞–º –¥–ª—è –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–µ–π.

```python
def add_features(df, max_lag, sizes):
    """–§—É–Ω–∫—Ü–∏—è –±–µ—Ä–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º, –∑–Ω–∞—á–µ–Ω–∏–µ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –ª–∞–≥–∞ 
    –∏ —Ä–∞–∑–º–µ—Ä—ã —Å–∫–æ–ª—å–∑—è—â–∏—Ö —Å—Ä–µ–¥–Ω–∏—Ö.
    –ù–∞ –≤—ã—Ö–æ–¥–µ –ø–æ–ª—É—á–∞–µ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ–º —ç—Ç–∏—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤."""
    
    for lag in range(1, max_lag + 1):
        df['lag_{}'.format(lag)] = df['num_orders'].shift(lag)
    
    for size in sizes:
        df['rolling_mean_{}'.format(size)] = df['num_orders'].shift().rolling(size).mean()
```

C–æ–∑–¥–∞–¥–∏–º –∫–æ–ø–∏—é –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞, —á—Ç–æ–±—ã –º–æ–∂–Ω–æ –±—ã–ª–æ —Å–ø–æ–∫–æ–π–Ω–æ –¥–æ–±–∞–≤–ª—è—Ç—å–≤ –Ω–µ–≥–æ –Ω–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏. –ü—Ä–∏–º–µ–Ω–∏–º —Ñ—É–Ω–∫—Ü–∏—é –¥–æ–±–∞–≤–∏–≤ –ª–∞–≥–∏ –¥–æ 24–≥–æ –∏ —Å–∫–æ–ª—å–∑—è—â–∏–µ —Å—Ä–µ–¥–Ω–∏–µ –∑–∞ –ø–æ–ª–¥–Ω—è, —Å—É—Ç–∫–∏, –¥–≤–∞ –¥–Ω—è, –Ω–µ–¥–µ–ª—é (12,24,48,168 —á–∞—Å–æ–≤). –¢–∞–∫ –∂–µ –¥–æ–±–∞–≤–∏–º 168 –ª–∞–≥, –ø–æ—Ç–æ–º—É —á—Ç–æ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –∞–Ω–∞–ª–æ–≥–∏—á–Ω–æ–≥–æ —á–∞—Å–∞ –Ω–∞ –ø—Ä–µ–¥—ã–¥—É—â–µ–π –Ω–µ–¥–µ–ª–∏ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –æ—á–µ–Ω—å –≤–∞–∂–Ω—ã–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–º, –ø–æ—Ç–æ–º—É —á—Ç–æ —É –Ω–∞—Å –∏–º–µ–µ—Ç—Å—è –Ω–µ–¥–µ–ª—å–Ω–∞—è –∏ —Å—É—Ç–æ—á–Ω–∞—è —Å–µ–∑–æ–Ω–Ω–æ—Å—å (–≤—Ä–µ–º—è —Å—É—Ç–æ–∫ –∏ –¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏).

–°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –æ–±—É–µ–Ω–∏—è, –ø–æ–¥–±–æ—Ä–∞ –ª—É—á—à–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è. –ë—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å TimeSeriesSplit() –ø—Ä–∏ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏, —á—Ç–æ–±—ã –Ω–µ –Ω–∞—Ä—É—à–∞–ª—Å—è –≤—Ä–µ–º–µ–Ω–Ω–æ–π —Ä—è–¥.

```python
tscv = TimeSeriesSplit(n_splits=8)
```

<div class="alert alert-block alert-info">
<b>–ò–∑–º–µ–Ω–µ–Ω–∏—è:</b> –ò—Å–ø—Ä–∞–≤–∏–ª, —Ç–µ–ø–µ—Ä—å –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤—Å—ë –≤–µ—Ä–Ω–æ )
</div>

<div class="alert alert-block alert-warning">
<b>‚ö†Ô∏è –°–æ–≤–µ—Ç:</b>

n_splits , –æ–±—ã—á–Ω–æ, –∑–∞–¥–∞—é—Ç —Ç–∞–∫, —á—Ç–æ–±—ã —Ä–∞–∑–º–µ—Ä –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ –ø—Ä–∏ –ø—Ä–æ–≤–µ–¥–µ–Ω–∏–∏ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏ –ø—Ä–∏–º–µ—Ä–Ω–æ —Å–æ–≤–ø–∞–¥–∞–ª —Å —Ä–∞–∑–º–µ—Ä–æ–º —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏. –†–∞–∑–º–µ—Ä —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏ —É –Ω–∞—Å 10%, –∞ —Ä–∞–∑–º–µ—Ä –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π —Ä–∞–≤–µ–Ω 1/(n_splits+1) —á–∞—Å—Ç–∏ –æ—Ç –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏ (—Å–º. –¥–æ–∫—É–º–µ–Ω—Ç–∞—Ü–∏—é –∫ TimeSeriesSplit). –ö –ø—Ä–∏–º–µ—Ä—É, –µ—Å–ª–∏ n_splits=2, —Ç–æ —Ä–∞–∑–º–µ—Ä –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π —Ä–∞–≤–µ–Ω 1/3 –æ—Ç –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏, —Ç.–µ. 1/3*90%=30% –æ—Ç –∏—Å—Ö–æ–¥–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏.
</div>

<div class="alert alert-block alert-success">
<b>V2 ‚úîÔ∏è –£—Å–ø–µ—à–Ω–æ –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–æ</b>
</div>

```python
def grid_fit(model,params):
    """–§—É–∫—Ü–∏—è –±–µ—Ä–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –º–æ–¥–µ–ª—å, –∏ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è GridSeachCV.
    –ü—Ä–æ–≤–æ–¥–∏—Ç –æ–±—É—á–µ–Ω–∏–µ —Å –≤—ã—è–≤–ª–µ–Ω–∏–µ–º –ª—É—á—à–∏—Ö –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –ø–æ –º–µ—Ç—Ä–∏–∫–µ RMSE.
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –æ–±—É—á–µ–Ω–Ω—É—é –º–æ–¥–µ–ª—å, RMSE, –≤—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏."""
  
    grid = GridSearchCV(model, 
                        param_grid=params, 
                        scoring='neg_root_mean_squared_error',
                        cv=tscv)
        
    grid.fit(X_train,y_train)
    
    i = (list(grid.cv_results_['rank_test_score'])).index(1)
    rmse = (grid.cv_results_['mean_test_score'][i]*-1)
    time_fit = grid.cv_results_['mean_fit_time'][i]
    time_pred = grid.cv_results_['mean_score_time'][i]
    
    print('–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –º–æ–¥–µ–ª–∏:', grid.best_params_)
    print('RMSE',rmse)
    print('–í—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏:', time_fit)
    print('–í—Ä–µ–º—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏:', time_pred)
    
    return grid, rmse, time_fit, time_pred
```

```python
def pred(model):
    """–§—É–Ω–∫—Ü–∏—è –±–µ—Ä—ë—Ç –Ω–∞ –≤—Ö–æ–¥ –º–æ–¥–µ–ª—å, –æ–±—É—á–∞–µ—Ç –µ—ë –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ,
    –¥–µ–ª–∞–µ—Ç –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –∏ –ø—Ä–æ–≤–µ—Ä—è–µ—Ç RMSE –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ.
    –ù–∞ –≤—ã—Ö–æ–¥ –∏–¥—ë—Ç RMSE, –≤—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è."""
    
    start = time.time()
    model.fit(X_train, y_train)
    end = time.time()
    time_fit = end - start

    start = time.time()
    preds = model.predict(X_test)
    end = time.time()
    rmse = mse(y_test, preds) ** 0.5
    time_pred = end - start
    
    print('–í—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏:', time_fit)
    print('–í—Ä–µ–º—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏:', time_pred)
    print('RMSE',rmse)
    
    return preds, rmse, time_fit, time_pred
```

–î–æ–±–∞–≤–∏–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç —Å–æ–µ–¥–µ–Ω—è—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ —Ç–∞–±–ª–∏—Ü—É.

```python
def metrics(df,model,score,fit,pred):
    '''–§—É–Ω–∫—Ü–∏—è –±–µ—Ä—ë—Ç –Ω–∞ –≤—Ö–æ–¥ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞, –Ω–∞–∑–≤–∞–Ω–∏–µ –º–æ–¥–µ–ª–∏, –ø–æ–∫–∞–∑–∞–µ—Ç–ª—å RMSE (–ø–æ CV),
    –≤—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è –∏ –≤—Ä–µ–º—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è. –î–æ–±–∞–≤–ª—è–µ—Ç —Å—Ç—Ä–æ–∫—É —Å –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è–º–∏ –º–æ–¥–µ–ª–∏.'''
    
    df.loc[model,'RMSE'] = RMSE
    df.loc[model,'–í—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è'] = fit
    df.loc[model,'–í—Ä–µ–º—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è'] = pred
    
    return df
```

–í —Ç–µ–æ—Ä–∏–∏ –≤ –¥–∞–Ω—ã—Ö –∑–∞–¥–∞—á–∞—Ö –¥–æ–ª–∂–Ω—ã –ª—É—á—à–µ —Ä–∞–±–æ—Ç–∞—Ç—å –ª–∏–Ω–µ–π–Ω—ã–µ –º–æ–¥–µ–ª–∏, –ø–æ—Ç–æ–º—É —á—Ç–æ –¥–µ—Ä–µ–≤—å—è —Ñ–∏–∑–∏—á–µ—Å–∫–∏ –Ω–µ –º–æ–≥—É—Ç –ø—Ä–µ–¥—Å–∫–∞–∑–∞—Ç—å –Ω–æ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è, –∞ –∫–∞–∫ –º—ã –≤–∏–¥–µ–ª–∏ –ø–æ —Ç—Ä–µ–Ω–¥—É - —Ü–µ–Ω–∞ –≤ —Å—Ä–µ–¥–Ω–µ–º –ø—Ä–æ–¥–æ–ª–∂–∞–µ—Ç —Ä–∞—Å—Ç–∏. –ù–æ –≤—Å—ë —Ä–∞–≤–Ω–æ—Å–Ω–∞—á–∞–ª–∞ –º—ã –ø–æ—Å–º–æ—Ç—Ä–∏–º –∫–∞–∫ –ø–æ–∫–∞–∂—É—Ç —Å–µ–±—è –æ–±—ã—á–Ω—ã–µ –º–æ–¥–µ–ª–∏, –∞ –∑–∞–¥–µ—Å—å –º–æ—Å–∑–¥–∞–¥–∏–º –≥–∏–±—Ä–∏–¥–Ω—É—é –º–æ–¥–µ–ª—å —Å –ø–æ–º–æ—â—å—é –±–∏–±–ª–µ–æ—Ç–µ–∫–∏ linear-tree.

#### LinearRegression

```python
pipe = Pipeline([('ohe_transformer', ohe_transformer),
                 ('lr', LR())])
params_lr = {'lr__normalize': [True,False]}

grid_lr, rmse_lr, lr_ft, lr_pt = grid_fit(pipe,params_lr)

metrics(results, "LinearRegression", rmse_lr, lr_ft, lr_pt)
```

#### Ridge

```python
pipe = Pipeline([('ohe_transformer', ohe_transformer),
                 ('ridge', Ridge())])

params_r = {'ridge__normalize': [True,False]}

grid_r, rmse_r, r_ft, r_pt = grid_fit(pipe,params_r)

metrics(results, "RidgeRegression", rmse_r, r_ft, r_pt)
```

–ü–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –∏–Ω–¥–µ–Ω—Ç–∏—á–Ω—ã–µ —Å –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–µ–π, —Ç–æ–ª—å–∫–æ —Å–∫–æ—Ä–æ—Å—Ç—å –æ–±—É—á–µ–Ω–∏—è –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –±—ã—Å—Ç—Ä–µ–µ.

#### CatBoostRegressor

```python
%%time

model_cb = CBR(loss_function='RMSE',
            random_state=s,
            early_stopping_rounds=100,
            verbose=1000, depth=6)

pipe = Pipeline([('oe_transformer', oe_transformer),
                 ('cb', model_cb)])

params_cb = {'cb__learning_rate':[0.01,0.015,0.02]}

grid_cb, rmse_cb, cb_ft, cb_pt = grid_fit(pipe,params_cb)

metrics(results, "CatBoostRegressor", rmse_cb, cb_ft, cb_pt)
```

#### LGBMRegressor

```python
%%time

model_lgbm = LGBMRegressor(random_state=s)

pipe = Pipeline([('oe_transformer', oe_transformer),
                 ('lgbm', model_lgbm)])

params_lgbm = [{'lgbm__num_leaves': [32], 
                'lgbm__learning_rate':[0.03,0.04]}]

lgbm_grid, rmse_lgbm, lgbm_time_fit, lgbm_time_pred = grid_fit(pipe,params=params_lgbm)

metrics(results, 'LGBMRegressor', rmse_lgbm, lgbm_time_fit, lgbm_time_pred)
```
