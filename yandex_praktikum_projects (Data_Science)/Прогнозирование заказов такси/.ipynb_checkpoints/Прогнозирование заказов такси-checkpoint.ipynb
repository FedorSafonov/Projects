## Обучение

Добавим день недели и час в исходные данные.

### Добавление признаков

```python
df['dayofweek'] = df.index.dayofweek
df['hour'] = df.index.hour
```

Создадим функцию для генерации признаков, которые понадобятся нам для обучения моделей.

```python
def add_features(df, max_lag, sizes):
    """Функция берет на вход датафрейм, значение максимального лага 
    и размеры скользящих средних.
    На выходе получаем датафрейм с добавлением этих признаков."""
    
    for lag in range(1, max_lag + 1):
        df['lag_{}'.format(lag)] = df['num_orders'].shift(lag)
    
    for size in sizes:
        df['rolling_mean_{}'.format(size)] = df['num_orders'].shift().rolling(size).mean()
```

Cоздадим копию датафрейма, чтобы можно было спокойно добавлятьв него новые признаки. Применим функцию добавив лаги до 24го и скользящие средние за полдня, сутки, два дня, неделю (12,24,48,168 часов). Так же добавим 168 лаг, потому что показатель аналогичного часа на предыдущей недели должен быть очень важным показателем, потому что у нас имеется недельная и суточная сезоннось (время суток и день недели).

```python
sizes = [12,24,48,168]
df1 = df.copy()

add_features(df1, 24, sizes)
df1['lag_168'] = df1['num_orders'].shift(168)
df1 = df1.dropna()
```

```python
df1.head(1).T
```

### Разбивка и функии

Разобьем данные без перемешивания и разделим выборки на целевой признак и остальные.

```python
train, test = tts(df1, shuffle=False, test_size=0.1)

X_train = train.drop(['num_orders'], axis=1)
y_train = train['num_orders']

X_test = test.drop(['num_orders'], axis=1)
y_test = test['num_orders']
```

```python
print('Размер признаков в обучабщей выборке:',X_train.shape[0])
print('Размер признаков в тестовой выборке:',X_test.shape[0])
print('Размер таргета в обучабщей выборке:',y_train.shape[0])
print('Размер таргета в тестовой выборке:',y_test.shape[0])
```

Так как у нас есть неочевидые категориальные признаки (день недели и час), то дабавим OHE и OE кодировщики, чтобы модели могли их адекватиние обработать.

```python
cat_features = ['dayofweek', 'hour']

ohe_transformer = make_column_transformer(
    (OneHotEncoder(), cat_features),
    remainder='passthrough')

oe_transformer = make_column_transformer(
    (OrdinalEncoder(), cat_features),
    remainder='passthrough')
```

Создадим функции для обуения, подбора лучших параметров и предсказания. Будем использовать TimeSeriesSplit() при кросс-валидации, чтобы не нарушался временной ряд.

```python
tscv = TimeSeriesSplit(n_splits=8)
```

```python
def grid_fit(model,params):
    """Фукция берет на вход модель, и гиперпараметры для GridSeachCV.
    Проводит обучение с выявлением лучших гиперпараметров по метрике RMSE.
    Возвращает обученную модель, RMSE, время обучения и предсказания модели."""
    
    grid = GridSearchCV(model, 
                        param_grid=params, 
                        scoring='neg_root_mean_squared_error',
                        cv=tscv)
        
    grid.fit(X_train,y_train)
    
    i = (list(grid.cv_results_['rank_test_score'])).index(1)
    rmse = (grid.cv_results_['mean_test_score'][i]*-1)
    time_fit = grid.cv_results_['mean_fit_time'][i]
    time_pred = grid.cv_results_['mean_score_time'][i]
    
    print('Лучшие параметры модели:', grid.best_params_)
    print('RMSE',rmse)
    print('Время обучения модели:', time_fit)
    print('Время предсказания модели:', time_pred)
    
    return grid, rmse, time_fit, time_pred
```

```python
def pred(model):
    """Функция берёт на вход модель, обучает её на обучающей выборке,
    делает предсказания и проверяет RMSE на тестовой выборке.
    На выход идёт RMSE, время обучения и предсказания."""
    
    start = time.time()
    model.fit(X_train, y_train)
    end = time.time()
    time_fit = end - start

    start = time.time()
    preds = model.predict(X_test)
    end = time.time()
    rmse = mse(y_test, preds) ** 0.5
    time_pred = end - start
    
    print('Время обучения модели:', time_fit)
    print('Время предсказания модели:', time_pred)
    print('RMSE',rmse)
    
    return preds, rmse, time_fit, time_pred
```

Добавим функцию, которая будет соеденять результат в таблицу.

```python
results = pd.DataFrame(columns=['RMSE','Время обучения','Время предсказания'])
```

```python
def metrics(df,model,score,fit,pred):
    '''Функция берёт на вход датафрейма, название модели, показаетль RMSE (по CV),
    время обучения и время предсказания. Добавляет строку с показателями модели.'''
    
    df.loc[model,'RMSE'] = score
    df.loc[model,'Время обучения'] = fit
    df.loc[model,'Время предсказания'] = pred
    
    return df
```

```python
# Добавляем опцию, чтобы все дробные числа в таблицах
# округлялись до второго знака после запятой.

pd.options.display.float_format = '{: .2f}'.format
```

### Стандартные модели

В теории в даных задачах должны лучше работать линейные модели, потому что деревья физически не могут предсказать новые значения, а как мы видели по тренду - цена в среднем продолжает расти. Но всё равно сначала мы посмотрим как покажут себя обычные модели, а задесь мосздадим гибридную модель с помощью библеотеки linear-tree.

#### LinearRegression

```python
pipe = Pipeline([('ohe_transformer', ohe_transformer),
                 ('lr', LR())])
params_lr = {'lr__normalize': [True,False]}

grid_lr, rmse_lr, lr_ft, lr_pt = grid_fit(pipe,params_lr)

metrics(results, "LinearRegression", rmse_lr, lr_ft, lr_pt)
```

#### Ridge

```python
pipe = Pipeline([('ohe_transformer', ohe_transformer),
                 ('ridge', Ridge())])

params_r = {'ridge__normalize': [True,False]}

grid_r, rmse_r, r_ft, r_pt = grid_fit(pipe,params_r)

metrics(results, "RidgeRegression", rmse_r, r_ft, r_pt)
```

Показатели индентичные с линейной регрессией, только скорость обучения и предсказания быстрее.

#### CatBoostRegressor

```python
%%time

model_cb = CBR(loss_function='RMSE',
            random_state=s,
            early_stopping_rounds=100,
            verbose=1000, depth=6)

pipe = Pipeline([('oe_transformer', oe_transformer),
                 ('cb', model_cb)])

params_cb = {'cb__learning_rate':[0.01,0.015,0.02]}

grid_cb, rmse_cb, cb_ft, cb_pt = grid_fit(pipe,params_cb)

metrics(results, "CatBoostRegressor", rmse_cb, cb_ft, cb_pt)
```

Модель обучается медленнее, и точность меньше.

#### LGBMRegressor

```python
%%time

model_lgbm = LGBMRegressor(random_state=s)

pipe = Pipeline([('oe_transformer', oe_transformer),
                 ('lgbm', model_lgbm)])

params_lgbm = [{'lgbm__num_leaves': [32], 
                'lgbm__learning_rate':[0.03,0.04]}]

lgbm_grid, rmse_lgbm, lgbm_time_fit, lgbm_time_pred = grid_fit(pipe,params=params_lgbm)

metrics(results, 'LGBMRegressor', rmse_lgbm, lgbm_time_fit, lgbm_time_pred)
```

Показывает себя хуже прочих моделей по RMSE.

```python
results.sort_values('RMSE')
```

Пока самыми удачными выходят линейные модели и в частности Ridge. 

### Гибридные модели

Бустинги имеют хороший потенциал, но  так как они основаны на решающем дереве, то не могут предсказывать будущие значение (значение, которого не было в выборке), а как мы видем временой ряд у нас не статический и имеет явный восходящий тренд. Это означает, что в будущем кличество заказов может менятся (вероятно увеличиваться) и обычные бустинги не смогут угадывать эти значения.

Дле решения этой задачи существуеют гибридные модели, которые соединяют в себе линейные и древовидные, в этом разделе мы опробуем их в деле.

#### LinearForestRegressor

```python
%%time

model_lfr = LFR(base_estimator=Ridge())

pipe = Pipeline([('ohe_transformer', ohe_transformer),
                 ('lfr', model_lfr)])

params_lfr = {'lfr__n_estimators': [100,150]}

lfr_grid, rmse_lfr, lfr_time_fit, lfr_time_pred = grid_fit(pipe, params_lfr)

metrics(results, 'LinearForestRegression', rmse_lfr, lfr_time_fit, lfr_time_pred)
```

#### LinearBoostRegressor

```python
%%time

model_lbr = LBR(base_estimator=Ridge())

pipe = Pipeline([('ohe_transformer', ohe_transformer),
                 ('lbr', model_lbr)])

params_lbr = {'lbr__max_depth' : [3,4,5],
                'lbr__n_estimators': [1,2,3,4,10]}

lbr_grid, rmse_lbr, lbr_time_fit, lbr_time_pred = grid_fit(pipe, params_lbr)

metrics(results, 'LinearBoostRegression', rmse_lbr, lbr_time_fit, lbr_time_pred)
```

```python
results.sort_values('RMSE')
```

По RSME незначительно выигрывает обычный CatBoostRegressor, но так как деревья не могут предсказывать будущее значения, то предпочтение отдаётся модели Ridge, которая показывает отличный результат и скорость её обучения в 100 раз быстрее чем у CatBoost.
